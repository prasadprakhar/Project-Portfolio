## Separate dataframes for the top 3 English speaking countries

```python
#dataframes of top3 english speaking and by raised amount usd --> refer Part 3 analysis
df1 = sectors[sectors['country_code']=='USA']
df2 = sectors[sectors['country_code']=='GBR']
df3 = sectors[sectors['country_code']=='IND']
```

```python
print(df1.shape,df2.shape,df3.shape)
```

```python
df1['raised_amount_usd'].describe()
```

```python
#select investment range of 5 million to 15 million USD for United States
df1 = df1[(df1['raised_amount_usd'] >= 5000000) & (df1['raised_amount_usd'] <= 15000000)]
```

```python
df1.shape
```

```python
#select investment range of 5 million to 15 million USD for GBR and India
df2 = df2[(df2['raised_amount_usd'] >= 5000000) & (df2['raised_amount_usd'] <= 15000000)]
df3 = df3[(df3['raised_amount_usd'] >= 5000000) & (df3['raised_amount_usd'] <= 15000000)]
```

```python
print(df1.shape,df2.shape,df3.shape)
```

```python
df3.info()
```

```python
# replace with the name of the main sectors
df1.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising']=df1.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising'].replace(1, pd.Series(df1.columns, df1.columns))
df2.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising']=df2.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising'].replace(1, pd.Series(df2.columns, df2.columns))
df3.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising']=df3.loc[:, 'Automotive & Sports':'Social, Finance, Analytics, Advertising'].replace(1, pd.Series(df3.columns, df3.columns))
```

```python
#aggregate all the text rows to arrive at a value for a new column 
df3['main_sector'] = df3[['Automotive & Sports', 'Cleantech / Semiconductors', 'Entertainment',
                         'Health', 'Manufacturing', 'News, Search and Messaging', 'Others',
                         'Social, Finance, Analytics, Advertising']].astype('str').agg(''.join, axis=1)
```

```python
#remove the digit 0 in the column main_sector
df3['main_sector'].replace('\d+', '', regex=True, inplace=True)
```
